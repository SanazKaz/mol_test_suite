```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
import matplotlib.ticker as mticker
from rdkit import Chem
from rdkit.Chem import QED

COLOR_PRISM = "orange"
COLOR_DIFF  = "steelblue"

# --- datasets (test proteins only, no 7E2Z) ---
datasets = {
    # 6CM4
    '6cm4_PRISM_QED_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/PRISM_test_pockets/PRISM_6cm4_30_nodes.sdf"
    ],
    '6cm4_DiffSBDD_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/DiffSBDD_test_pockets/DiffSBDD_6cm4_30_nodes.sdf"
    ],
    # 6LUQ
    '6luq_PRISM_QED_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/PRISM_test_pockets/PRISM_6luq_30_nodes.sdf"
    ],
    '6luq_DiffSBDD_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/DiffSBDD_test_pockets/DiffSBDD_6luq_test_30_nodes.sdf"
    ]
}

# ---------- Statistical Helper Functions (from SuCOS analysis) ----------
def _as_clean_array(x):
    """Clean array by removing non-finite values"""
    x = np.asarray(x, dtype=float)
    return x[np.isfinite(x)]

def cliffs_delta(x, y):
    """
    Calculate Cliff's delta effect size measure.
    
    Args:
        x, y: Arrays of values to compare
        
    Returns:
        float: Cliff's delta (-1 to 1, where 0 = no effect)
    """
    x = _as_clean_array(x)
    y = _as_clean_array(y)
    nx, ny = len(x), len(y)
    if nx == 0 or ny == 0: 
        return np.nan
    if nx * ny <= 100000:
        diff = x[:, None] - y[None, :]
        return (np.sum(diff > 0) - np.sum(diff < 0)) / (nx * ny)
    # For large datasets, use sampling
    rng = np.random.default_rng(123)
    xs = rng.choice(x, size=min(nx, 500), replace=False)
    ys = rng.choice(y, size=min(ny, 500), replace=False)
    diff = xs[:, None] - ys[None, :]
    return (np.sum(diff > 0) - np.sum(diff < 0)) / (xs.size * ys.size)

def run_stats(x, y, label):
    """
    Run statistical comparison between two groups.
    
    Args:
        x, y: Arrays of values to compare
        label: String label for the comparison
        
    Returns:
        tuple: (p_value, cliff_delta)
    """
    print(f"\n=== {label}: PRISM vs DiffSBDD ===")
    x = _as_clean_array(x)
    y = _as_clean_array(y)
    print(f"n_PRISM={len(x)}, n_DiffSBDD={len(y)}")
    u = mannwhitneyu(x, y, alternative="two-sided")
    delta = cliffs_delta(x, y)
    print(f"Mann–Whitney U: U={u.statistic:.4g}, p={u.pvalue:.3g}")
    print(f"Cliff's δ={delta:.3f} (|δ|≈0.147 small, 0.33 medium, 0.474 large)")
    return u.pvalue, delta

def p_to_stars(p):
    """Convert p-value to significance stars"""
    if p <= 0.0001: return "****"
    elif p <= 0.001: return "***"
    elif p <= 0.01: return "**"
    elif p <= 0.05: return "*"
    else: return "ns"

def violin_with_stats(ax, data, labels, ylabel, pval=None, delta=None):
    """
    Create violin plot with statistical annotations.
    
    Args:
        ax: Matplotlib axis
        data: List of arrays for each group
        labels: List of labels for groups
        ylabel: Y-axis label
        pval: P-value for significance stars
        delta: Cliff's delta for effect size
    """
    parts = ax.violinplot(data, showmeans=True, showmedians=True)
    for pc, color in zip(parts['bodies'], [COLOR_PRISM, COLOR_DIFF]):
        pc.set_facecolor(color)
        pc.set_alpha(0.65)
    # Make means/medians easier to see (matching your original style)
    if 'cmeans' in parts:
        parts['cmeans'].set_linewidth(1.2)
    if 'cmedians' in parts:
        parts['cmedians'].set_linewidth(1.2)
        
    ax.set_xticks([1, 2])
    ax.set_xticklabels(labels)
    ax.set_ylabel(ylabel)
    ax.set_ylim(0, 1.0)
    ax.set_yticks(np.arange(0, 1.1, 0.2))
    
    ax.grid(False)  # No grid like your original

    # Add significance stars and effect size with smaller fonts
    if pval is not None:
        ymin, ymax = ax.get_ylim()
        # Expand ylim for headroom
        ax.set_ylim(ymin, ymax + 0.15*(ymax - ymin))
        ymin, ymax = ax.get_ylim()
        
        # Place stars above violins
        y_star = ymax - 0.1*(ymax - ymin)
        ax.plot([1, 2], [y_star, y_star], color="black", lw=1.2)
        ax.text(
            1.5,
            y_star + 0.01*(ymax - ymin),
            p_to_stars(pval),
            ha="center", va="bottom",
            fontsize=14
        )

    if delta is not None:
        # Add effect size in top-right corner
        ax.text(
            2.3,
            ymax - 0.02*(ymax - ymin),
            f"δ={delta:.3f}",
            ha="right", va="top",
            fontsize=14
        )

def two_hist(ax, prism_vals, diff_vals, xlabel, bins=20):
    """Create overlapping histograms for two groups (kept for potential future use)"""
    ax.hist(prism_vals, bins=bins, alpha=0.6, color=COLOR_PRISM, label="PRISM")
    ax.hist(diff_vals,  bins=bins, alpha=0.6, color=COLOR_DIFF,  label="DiffSBDD")
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Count")
    ax.legend(frameon=False)
    ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True))
    ax.set_ylim(0, 1.0)
    ax.set_yticks(np.arange(0, 1.1, 0.1))

# --- helper to load SDF and compute QED ---
def load_sdf_and_qed(path, tag):
    """Load molecules from SDF and calculate QED scores"""
    mols = []
    try:
        mols = [m for m in Chem.SDMolSupplier(path) if m is not None]
    except Exception as e:
        print(f"[WARN] Failed to read: {path} — {e}")
        return []
    return [(tag, os.path.basename(path), QED.qed(m)) for m in mols if m is not None]

# --- Load and process data ---
rows = []
for tag, files in datasets.items():
    for f in files:
        rows += load_sdf_and_qed(f, tag)

df = pd.DataFrame(rows, columns=["set","file","QED"])

# --- tidy labels (PRISM vs DiffSBDD + pocket) ---
def pocket_method(tag):
    """Extract pocket ID and method from tag"""
    p = tag.split("_")[0]           # pocket
    m = "PRISM" if "PRISM" in tag else "DiffSBDD"
    return p, m

df[["pocket","method"]] = df["set"].apply(lambda t: pd.Series(pocket_method(t)))

# --- print counts per pocket/method and totals ---
print("\nCounts per pocket/method:")
counts = df.groupby(["pocket","method"]).size().reset_index(name="n")
for _, r in counts.iterrows():
    print(f"  {r['pocket'].upper():>4}  {r['method']:<8}  n = {int(r['n'])}")
print(f"\nTOTAL molecules: {len(df)}")

# Save counts and per-molecule data
counts.to_csv("qed_counts_per_pocket_method.csv", index=False)
df.to_csv("qed_per_molecule_values.csv", index=False)

# ---------- Statistical Analysis by Pocket ----------
pockets = ["6cm4", "6luq"]
stats_results = {}

for pocket in pockets:
    prism_vals = df[(df["pocket"]==pocket) & (df["method"]=="PRISM")]["QED"].values
    diff_vals  = df[(df["pocket"]==pocket) & (df["method"]=="DiffSBDD")]["QED"].values
    
    p_val, delta = run_stats(prism_vals, diff_vals, f"QED {pocket.upper()}")
    stats_results[pocket] = {'p_value': p_val, 'cliff_delta': delta}

# ---------- Create Plots ----------
# Side-by-side violin plots for all pockets (matching your original style)
fig, axes = plt.subplots(1, 2, figsize=(10,5), sharey=False, dpi=300)

for ax, pocket in zip(axes, pockets):
    prism_vals = df[(df["pocket"]==pocket) & (df["method"]=="PRISM")]["QED"].values
    diff_vals  = df[(df["pocket"]==pocket) & (df["method"]=="DiffSBDD")]["QED"].values
    
    violin_with_stats(
        ax, 
        [prism_vals, diff_vals], 
        ["PRISM","DiffSBDD"], 
        "QED",
        stats_results[pocket]['p_value'],
        stats_results[pocket]['cliff_delta']
    )
    
    n_prism = len(prism_vals)
    n_diff  = len(diff_vals)
    ax.set_title(f"{pocket.upper()}", fontsize=14)

plt.tight_layout(pad=3.0)
fig.savefig("/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/figures/qed_analysis_gpcr_test_pockets_violins**.png", dpi=300, bbox_inches="tight")
# fig.savefig("qed_analysis_all_pockets_violins.svg", bbox_inches="tight")
plt.show()

print("\n" + "="*50)
print("STATISTICAL SUMMARY")
print("="*50)
for pocket in pockets:
    p_val = stats_results[pocket]['p_value']
    delta = stats_results[pocket]['cliff_delta']
    significance = p_to_stars(p_val)
    print(f"{pocket.upper()}: p={p_val:.3g} ({significance}), δ={delta:.3f}")

print("\nFiles saved:")
print(" - qed_analysis_all_pockets_violins.png")
print(" - qed_counts_per_pocket_method.csv")
print(" - qed_per_molecule_values.csv")
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
import matplotlib.ticker as mticker
from rdkit import Chem
from rdkit.Chem import QED

COLOR_PRISM = "orange"
COLOR_DIFF  = "steelblue"

# --- Single pocket dataset - 7E2Z ---
datasets = {
    '7e2z_PRISM_QED_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/PRISM_test_pockets/PRISM_7e2z_30_nodes.sdf"
    ],
    '7e2z_DiffSBDD_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/DiffSBDD_test_pockets/DiffSBDD_7e2z_30_nodes.sdf"
    ]
}

# ---------- Statistical Helper Functions ----------
def _as_clean_array(x):
    """Clean array by removing non-finite values"""
    x = np.asarray(x, dtype=float)
    return x[np.isfinite(x)]

def cliffs_delta(x, y):
    """
    Calculate Cliff's delta effect size measure.
    
    Args:
        x, y: Arrays of values to compare
        
    Returns:
        float: Cliff's delta (-1 to 1, where 0 = no effect)
    """
    x = _as_clean_array(x)
    y = _as_clean_array(y)
    nx, ny = len(x), len(y)
    if nx == 0 or ny == 0: 
        return np.nan
    if nx * ny <= 100000:
        diff = x[:, None] - y[None, :]
        return (np.sum(diff > 0) - np.sum(diff < 0)) / (nx * ny)
    # For large datasets, use sampling
    rng = np.random.default_rng(123)
    xs = rng.choice(x, size=min(nx, 500), replace=False)
    ys = rng.choice(y, size=min(ny, 500), replace=False)
    diff = xs[:, None] - ys[None, :]
    return (np.sum(diff > 0) - np.sum(diff < 0)) / (xs.size * ys.size)

def run_stats(x, y, label):
    """
    Run statistical comparison between two groups.
    
    Args:
        x, y: Arrays of values to compare
        label: String label for the comparison
        
    Returns:
        tuple: (p_value, cliff_delta)
    """
    print(f"\n=== {label}: PRISM vs DiffSBDD ===")
    x = _as_clean_array(x)
    y = _as_clean_array(y)
    print(f"n_PRISM={len(x)}, n_DiffSBDD={len(y)}")
    u = mannwhitneyu(x, y, alternative="two-sided")
    delta = cliffs_delta(x, y)
    print(f"Mann–Whitney U: U={u.statistic:.4g}, p={u.pvalue:.3g}")
    print(f"Cliff's δ={delta:.3f} (|δ|≈0.147 small, 0.33 medium, 0.474 large)")
    return u.pvalue, delta

def p_to_stars(p):
    """Convert p-value to significance stars"""
    if p <= 0.0001: return "****"
    elif p <= 0.001: return "***"
    elif p <= 0.01: return "**"
    elif p <= 0.05: return "*"
    else: return "ns"

def violin_with_stats(ax, data, labels, ylabel, pval=None, delta=None):
    """
    Create violin plot with statistical annotations.
    
    Args:
        ax: Matplotlib axis
        data: List of arrays for each group
        labels: List of labels for groups
        ylabel: Y-axis label
        pval: P-value for significance stars
        delta: Cliff's delta for effect size
    """
    parts = ax.violinplot(data, showmeans=True, showmedians=True)
    for pc, color in zip(parts['bodies'], [COLOR_PRISM, COLOR_DIFF]):
        pc.set_facecolor(color)
        pc.set_alpha(0.65)
    # Make means/medians easier to see
    if 'cmeans' in parts:
        parts['cmeans'].set_linewidth(1.2)
    if 'cmedians' in parts:
        parts['cmedians'].set_linewidth(1.2)
        
    ax.set_xticks([1, 2])
    ax.set_xticklabels(labels)
    ax.set_ylabel(ylabel)
    ax.set_ylim(0, 1.0)
    ax.set_yticks(np.arange(0, 1.1, 0.2))
    ax.grid(False)

    # Add significance stars and effect size
    if pval is not None:
        ymin, ymax = ax.get_ylim()
        # Expand ylim for headroom
        ax.set_ylim(ymin, ymax + 0.15*(ymax - ymin))
        ymin, ymax = ax.get_ylim()
        
        # Place stars above violins
        y_star = ymax - 0.1*(ymax - ymin)
        ax.plot([1, 2], [y_star, y_star], color="black", lw=1.2)
        ax.text(
            1.5,
            y_star + 0.01*(ymax - ymin),
            p_to_stars(pval),
            ha="center", va="bottom",
            fontsize=14
        )

    if delta is not None:
        # Add effect size in top-right corner
        ax.text(
            2.3,
            ymax - 0.02*(ymax - ymin),
            f"δ={delta:.3f}",
            ha="right", va="top",
            fontsize=14
        )

# --- helper to load SDF and compute QED ---
def load_sdf_and_qed(path, tag):
    """Load molecules from SDF and calculate QED scores"""
    mols = []
    try:
        mols = [m for m in Chem.SDMolSupplier(path) if m is not None]
    except Exception as e:
        print(f"[WARN] Failed to read: {path} — {e}")
        return []
    return [(tag, os.path.basename(path), QED.qed(m)) for m in mols if m is not None]

# --- Load and process data ---
rows = []
for tag, files in datasets.items():
    for f in files:
        rows += load_sdf_and_qed(f, tag)

df = pd.DataFrame(rows, columns=["set","file","QED"])

# --- tidy labels (PRISM vs DiffSBDD + pocket) ---
def pocket_method(tag):
    """Extract pocket ID and method from tag"""
    p = tag.split("_")[0]           # pocket
    m = "PRISM" if "PRISM" in tag else "DiffSBDD"
    return p, m

df[["pocket","method"]] = df["set"].apply(lambda t: pd.Series(pocket_method(t)))

# --- print counts per method and totals ---
print("\nCounts per method:")
counts = df.groupby(["method"]).size().reset_index(name="n")
for _, r in counts.iterrows():
    print(f"  {r['method']:<8}  n = {int(r['n'])}")
print(f"\nTOTAL molecules: {len(df)}")

# Save counts and per-molecule data
counts.to_csv("qed_7e2z_counts_per_method.csv", index=False)
df.to_csv("qed_7e2z_per_molecule_values.csv", index=False)

# ---------- Statistical Analysis ----------
prism_vals = df[df["method"]=="PRISM"]["QED"].values
diff_vals  = df[df["method"]=="DiffSBDD"]["QED"].values

p_val, delta = run_stats(prism_vals, diff_vals, "QED 7E2Z")

# ---------- Create Single Plot ----------
fig, ax = plt.subplots(1, 1, figsize=(6, 5), dpi=300)

violin_with_stats(
    ax, 
    [prism_vals, diff_vals], 
    ["PRISM","DiffSBDD"], 
    "QED",
    p_val,
    delta
)

ax.set_title("7E2Z", fontsize=14)

plt.tight_layout()
fig.savefig("/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/figures/qed_analysis_7e2z_violin.png", dpi=300, bbox_inches="tight")
plt.show()

print("\n" + "="*50)
print("STATISTICAL SUMMARY")
print("="*50)
significance = p_to_stars(p_val)
print(f"7E2Z: p={p_val:.3g} ({significance}), δ={delta:.3f}")

print("\nFiles saved:")
print(" - qed_analysis_7e2z_violin.png")
print(" - qed_7e2z_counts_per_method.csv")
print(" - qed_7e2z_per_molecule_values.csv")
```

```python
import os, random
from rdkit import Chem
from rdkit.Chem import Draw, AllChem, QED, Lipinski
from IPython.display import display
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import io
from PIL import Image

# ── helper to read all molecules from SDFs ──
def mols_from_sdfs(sdf_file_paths):
    mols = []
    for sdf_path in sdf_file_paths:
        if not os.path.exists(sdf_path):
            print(f"⚠️ File not found, skipping: {sdf_path}")
            continue
        try:
            suppl = Chem.SDMolSupplier(sdf_path)
            mols.extend([m for m in suppl if m is not None])
        except Exception as e:
            print(f"❌ Error reading {sdf_path}: {e}")
    return mols

def legend_for(m):
    try:
        qed = QED.qed(m)
    except Exception:
        qed = float("nan")
    rotb = Lipinski.NumRotatableBonds(m)
    arom = Lipinski.NumAromaticRings(m)
    return f"QED {qed:.3f}  |  RotB {rotb}  |  Arom {arom}"

# ── Reference molecule ──
reference_smiles = "C1CC(=O)NC2=C1C=CC(=C2)OCCCCN3CCN(CC3)C4=C(C(=CC=C4)Cl)Cl"
reference_mol = Chem.MolFromSmiles(reference_smiles)
if reference_mol:
    AllChem.Compute2DCoords(reference_mol)

# ── Updated datasets: 6CM4 and 6V0U ──
datasets = {
    '6cm4_PRISM_QED_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/PRISM_test_pockets/PRISM_6cm4_30_nodes.sdf"
    ],
    '6cm4_DiffSBDD_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/DiffSBDD_test_pockets/DiffSBDD_6cm4_30_nodes.sdf"
    ],
    # 6LUQ
    '6luq_PRISM_QED_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/PRISM_test_pockets/PRISM_6luq_30_nodes.sdf"
    ],
    '6luq_DiffSBDD_30_nodes': [
        "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/DiffSBDD_test_pockets/DiffSBDD_6luq_test_30_nodes.sdf"
    ]
}

datasets_to_display = [
    '6cm4_PRISM_QED_30_nodes',
    '6cm4_DiffSBDD_30_nodes',
    '6luq_PRISM_QED_30_nodes',
    '6luq_DiffSBDD_30_nodes'
]

# ── render settings with larger fonts ──
BASE_DIRECTORY = os.getcwd()
output_dir = os.path.join(BASE_DIRECTORY, "qed_random_samples_6cm4_6v0u")
os.makedirs(output_dir, exist_ok=True)

random.seed(100)  # reproducible sampling

# First display reference molecule
if reference_mol:
    print("\n=== REFERENCE MOLECULE ===")
    ref_legend = f"Aripiprazole: {legend_for(reference_mol)}"
    
    ref_img_data = Draw.MolsToGridImage(
        [reference_mol],
        molsPerRow=1,
        subImgSize=(500, 500),
        legends=[ref_legend],
        useSVG=False,
        returnPNG=True
    )
    
    # Convert to PIL and display with matplotlib for bigger text
    ref_img = Image.open(io.BytesIO(ref_img_data.data))
    
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.imshow(ref_img)
    ax.set_title('Reference Molecule', fontsize=18, fontweight='bold', pad=20)
    ax.axis('off')
    
    plt.tight_layout()
    ref_filepath = os.path.join(output_dir, "reference_molecule.png")
    plt.savefig(ref_filepath, dpi=300, bbox_inches='tight')
    print(f"✅ Saved: {ref_filepath}")
    plt.show()

# Then display dataset samples
for dataset_name in datasets_to_display:
    sdf_paths = [os.path.join(BASE_DIRECTORY, p) for p in datasets[dataset_name]]
    mols = mols_from_sdfs(sdf_paths)
    if not mols:
        print(f"No molecules found for {dataset_name}")
        continue

    # sample 9 random molecules + add reference molecule
    k = min(9, len(mols))
    sample_mols = random.sample(mols, k)
    
    # Add reference molecule at the beginning
    if reference_mol:
        all_mols = [reference_mol] + sample_mols
        ref_legend = f"reference: {legend_for(reference_mol)}"
        sample_legends = [legend_for(m) for m in sample_mols]
        legends = [ref_legend] + sample_legends
    else:
        all_mols = sample_mols
        legends = [legend_for(m) for m in sample_mols]

    # compute 2D coords for clean depictions
    for m in all_mols:
        try:
            AllChem.Compute2DCoords(m)
        except Exception:
            pass
    
    # Parse pocket and method from dataset name
    parts = dataset_name.split('_')
    pocket = parts[0].lower()  # e.g., "6CM4"
    method = parts[1]          # e.g., "PRISM" or "DiffSBDD"

    title = f"Random samples for {pocket} by {method}"

    # Create grid image
    img_data = Draw.MolsToGridImage(
        all_mols,
        molsPerRow=5,
        subImgSize=(600, 600),
        legends=legends,
        useSVG=False,
        returnPNG=True
    )

    # Convert to PIL and display with matplotlib for bigger text
    img = Image.open(io.BytesIO(img_data.data))
    
    fig, ax = plt.subplots(figsize=(15, 8))
    ax.imshow(img)
    ax.set_title(title, fontsize=18, fontweight='bold', pad=20)
    ax.axis('off')
    
    plt.tight_layout()
    
    # save
    filepath = os.path.join(output_dir, f"{dataset_name}_random.png")
    plt.savefig(filepath, dpi=300, bbox_inches='tight')
    print(f"✅ Saved: {filepath}")
    plt.show()
```
