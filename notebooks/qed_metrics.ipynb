```python
import argparse
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from rdkit import Chem
from rdkit.Chem import QED

# ==============================================================================
# ---  1. CONFIGURATION SECTION (Edit here) ---
# ==============================================================================

# Define the pockets you want to generate subplots for.
POCKETS_TO_PLOT = ["6CM4", "6LUQ", "6V0U"]

# Define the methods to compare. The path can be to a SINGLE .sdf file
# OR to a DIRECTORY containing multiple .sdf files.
METHODS_TO_COMPARE = {
    "Method A": "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/DiffSBDD_test_pockets", # <-- CHANGE THIS
    "Method B": "/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_sigmoid/DiffSBDD_test_pockets",      # <-- CHANGE THIS
}

# Define where the final plot will be saved.
OUTPUT_FILE = "test_Delete.png"

# ==============================================================================
# ---  2. SCRIPT LOGIC (No need to edit below) ---
# ==============================================================================

# ---------- Styling ----------
def _set_pub_style():
    """Sets a clean, publication-ready style for the plot."""
    sns.set_theme(context="talk", style="whitegrid")
    plt.rcParams.update({
        "figure.dpi": 150, "savefig.dpi": 300, "axes.titleweight": "bold",
        "axes.labelweight": "semibold", "axes.labelsize": 14,
        "axes.titlesize": 16, "xtick.labelsize": 12, "ytick.labelsize": 12,
        "legend.title_fontsize": 12, "legend.fontsize": 11,
        "axes.spines.top": False, "axes.spines.right": False,
        "grid.linestyle": "--", "grid.linewidth": 0.5,
    })

def plot_qed_for_pockets():
    """
    Loads molecules, filters by pocket based on filename, calculates QED, 
    and generates a multi-plot figure comparing distributions.
    """
    _set_pub_style()

    # --- Load Molecules and Calculate QED ---
    all_qed_data = []
    stats = {name: {"total_read": 0, "failed_sanitize": 0, "processed": 0} for name in METHODS_TO_COMPARE}
    
    print("--- Processing Files ---")
    for method_name, path_str in METHODS_TO_COMPARE.items():
        input_path = Path(path_str)
        files_to_process = []

        if not input_path.exists():
            print(f"Warning: Path does not exist for method '{method_name}': {input_path}")
            continue

        if input_path.is_dir():
            files_to_process.extend(sorted(input_path.glob("*.sdf")))
        elif input_path.is_file() and input_path.suffix == '.sdf':
            files_to_process.append(input_path)
        
        if not files_to_process:
            print(f"Warning: No .sdf files found for method '{method_name}' at path '{input_path}'")
            continue

        print(f"Processing {len(files_to_process)} file(s) for method '{method_name}'...")
        for sdf_path in files_to_process:
            # --- New Logic: Determine pocket from filename ---
            pocket_for_this_file = None
            filename_lower = sdf_path.name.lower()
            for target_pocket in POCKETS_TO_PLOT:
                if target_pocket.lower() in filename_lower:
                    pocket_for_this_file = target_pocket
                    break
            
            if not pocket_for_this_file:
                # print(f"  - Skipping file (no matching pocket in name): {sdf_path.name}")
                continue

            # Load molecules without sanitizing immediately to handle errors gracefully
            supplier = Chem.SDMolSupplier(str(sdf_path), removeHs=False, sanitize=False)
            for mol in supplier:
                if mol is None:
                    continue
                
                stats[method_name]["total_read"] += 1
                
                # Try to sanitize the molecule now
                try:
                    Chem.SanitizeMol(mol)
                except Exception:
                    stats[method_name]["failed_sanitize"] += 1
                    continue # Skip to the next molecule if sanitization fails

                # If sanitization succeeds, calculate QED and store data
                qed_score = QED.qed(mol)
                all_qed_data.append({
                    "Method": method_name,
                    "QED": qed_score,
                    "Pocket": pocket_for_this_file
                })
                stats[method_name]["processed"] += 1

    # --- Helpful Debugging Information ---
    print("\n--- Data Loading Summary ---")
    for method, s in stats.items():
        print(f"  Method: {method}")
        print(f"    - Total molecules read: {s['total_read']:,}")
        print(f"    - Failed sanitization (invalid chem): {s['failed_sanitize']:,}")
        print(f"    - Molecules successfully processed: {s['processed']:,}")

    if not all_qed_data:
        print("\nError: No valid molecules matched the specified pockets. No plot generated.")
        return

    df = pd.DataFrame(all_qed_data)

    # --- Create Subplots ---
    num_pockets = len(POCKETS_TO_PLOT)
    fig, axes = plt.subplots(
        1, num_pockets,
        figsize=(5 * num_pockets, 5),
        sharey=True
    )
    if num_pockets == 1:
        axes = [axes]

    print("\n--- Generating Subplots ---")
    for i, pocket_name in enumerate(POCKETS_TO_PLOT):
        ax = axes[i]
        pocket_df = df[df["Pocket"] == pocket_name]

        if pocket_df.empty:
            print(f"Warning: No data for pocket '{pocket_name}'. Subplot will be empty.")
            ax.text(0.5, 0.5, f"No data for\n{pocket_name}", ha='center', va='center', style='italic')
            ax.set_xticks([]); ax.set_yticks([])
            continue

        sns.kdeplot(
            data=pocket_df, x="QED", hue="Method",
            ax=ax, fill=True, alpha=0.2, linewidth=2.5,
        )
        ax.set_title(f"Pocket: {pocket_name}", fontsize=14)
        ax.set_xlabel("QED Score")
        ax.set_xlim(0, 1)
        if ax.get_legend():
            ax.get_legend().remove()

    axes[0].set_ylabel("Density")

    # --- Finalize and Save the Plot ---
    handles, labels = axes[0].get_legend_handles_labels()
    fig.legend(handles, labels, title="Method", loc="upper right", bbox_to_anchor=(0.99, 0.95))
    fig.suptitle("QED Distribution by Method and Pocket", fontsize=18, y=1.02)

    plt.tight_layout(rect=[0, 0, 0.9, 1])

    output_path = Path(OUTPUT_FILE)
    output_path.parent.mkdir(parents=True, exist_ok=True)
    plt.savefig(output_path)
    svg_path = output_path.with_suffix(".svg")
    plt.savefig(svg_path)
    print(f"\n Plot saved to: {output_path}")
    print(f" Vector SVG saved to: {svg_path}")

def main():
    """Main entry point to run the analysis."""
    plot_qed_for_pockets()

if __name__ == "__main__":
    main()


```

```python
import os, sys, numpy as np, pandas as pd
from rdkit import Chem, DataStructs
from rdkit.Chem import QED, Crippen, Descriptors, Lipinski, AllChem
from rdkit.Chem import RDConfig

# SA Score (RDKit contrib)
sys.path.append(os.path.join(RDConfig.RDContribDir, "SA_Score"))
import sascorer

# ------------------------
# Helpers
# ------------------------
def load_sdf_list(paths):
    mols = []
    for p in sorted(set(paths)):  # dedupe paths
        try:
            mols.extend([m for m in Chem.SDMolSupplier(p) if m is not None])
        except Exception as e:
            print(f"[WARN] Failed to read {p}: {e}")
    return mols

def morgan_fp(smiles_list, nBits=2048, radius=2):
    fps = []
    for s in smiles_list:
        m = Chem.MolFromSmiles(s)
        if m is None: 
            continue
        fps.append(AllChem.GetMorganFingerprintAsBitVect(m, radius, nBits=nBits))
    return fps

def avg_pairwise_dissimilarity(fps):
    n = len(fps)
    if n < 2:
        return np.nan
    vals = []
    for i in range(n):
        for j in range(i+1, n):
            vals.append(1.0 - DataStructs.TanimotoSimilarity(fps[i], fps[j]))
    return float(np.mean(vals)) if vals else np.nan

def lipinski_rules(m):
    """Return number of rules satisfied out of 5:
       MW  500, HBD  5, HBA  10, logP  5, rotatable bonds  10."""
    mw   = Descriptors.MolWt(m)
    hbd  = Lipinski.NumHDonors(m)
    hba  = Lipinski.NumHAcceptors(m)
    logp = Crippen.MolLogP(m)
    rot  = Lipinski.NumRotatableBonds(m)
    return int((mw <= 500) + (hbd <= 5) + (hba <= 10) + (logp <= 5) + (rot <= 10))

def per_mol_props(m):
    try:
        smi = Chem.MolToSmiles(m, canonical=True)
        return dict(
            smiles=smi,
            QED=QED.qed(m),
            SA=sascorer.calculateScore(m),
            logP=Crippen.MolLogP(m),
            LipinskiCount=lipinski_rules(m)
        )
    except Exception:
        return None

def props_from_paths(paths):
    mols = load_sdf_list(paths)
    rows = []
    for m in mols:
        r = per_mol_props(m)
        if r: rows.append(r)
    df = pd.DataFrame(rows)
    return df

def summarize_set(df):
    if df.empty:
        return dict(n_mols=0, QED_mean=np.nan, QED_median=np.nan,
                    SA_mean=np.nan, SA_median=np.nan,
                    logP_mean=np.nan, logP_median=np.nan,
                    Lipinski_mean=np.nan, Lipinski_median=np.nan,
                    diversity=np.nan)
    # diversity on canonical SMILES
    fps = morgan_fp(df["smiles"].tolist(), nBits=2048, radius=2)
    return dict(
        n_mols=int(len(df)),
        QED_mean=float(df["QED"].mean()),
        QED_median=float(df["QED"].median()),
        SA_mean=float(df["SA"].mean()),
        SA_median=float(df["SA"].median()),
        logP_mean=float(df["logP"].mean()),
        logP_median=float(df["logP"].median()),
        Lipinski_mean=float(df["LipinskiCount"].mean()),
        Lipinski_median=float(df["LipinskiCount"].median()),
        diversity=float(avg_pairwise_dissimilarity(fps))
    )

# ------------------------
# Collapse datasets into method buckets and pocketmethod buckets
# ------------------------
# Use your existing `datasets` dict, e.g. pasted above
# datasets = { ... }

# Method-level across all pockets and node sizes
methods = {"PRISM": [], "DiffSBDD": []}
for tag, files in datasets.items():
    (methods["PRISM"] if "PRISM" in tag else methods["DiffSBDD"]).extend(files)

# Pocket  method
by_pocket_method = {}
for tag, files in datasets.items():
    pocket = tag.split("_")[0].lower()
    method = "PRISM" if "PRISM" in tag else "DiffSBDD"
    by_pocket_method.setdefault((pocket, method), []).extend(files)

# ------------------------
# Compute summaries
# ------------------------
# Method-level
method_rows = []
for method, paths in methods.items():
    dfm = props_from_paths(paths)
    summ = summarize_set(dfm)
    summ.update(method=method)
    method_rows.append(summ)
    # Save per-molecule metrics per method in case you need them later
    dfm.to_csv(f"props_per_molecule_{method}.csv", index=False)

method_summary = pd.DataFrame(method_rows).set_index("method").round(3)
print("\nMethod-level summary (all pockets combined):")
print(method_summary)
method_summary.to_csv("method_level_metrics_summary.csv")

# Pocket  method
pm_rows = []
for (pocket, method), paths in sorted(by_pocket_method.items()):
    dfm = props_from_paths(paths)
    summ = summarize_set(dfm)
    summ.update(pocket=pocket.upper(), method=method)
    pm_rows.append(summ)
    # Save per-molecule metrics for this pocketmethod
    dfm.to_csv(f"props_per_molecule_{pocket}_{method}.csv", index=False)

pm_summary = pd.DataFrame(pm_rows).sort_values(["pocket","method"]).round(3)
print("\nPer pocket  method summary:")
print(pm_summary)
pm_summary.to_csv("per_pocket_method_metrics_summary.csv", index=False)
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import QED
import os
from typing import List
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- Part 1: Data Processing ---

def calculate_qed_for_sdf(sdf_file_path: str) -> List[float]:
    """
    Calculate QED scores for all molecules in a single SDF file.
    
    Args:
        sdf_file_path (str): The full path to the SDF file.

    Returns:
        A list of QED scores for valid molecules.
    """
    qed_scores = []
    if not os.path.exists(sdf_file_path):
        print(f" Warning: File not found, skipping: {sdf_file_path}")
        return qed_scores

    try:
        # Use a context manager for safer file handling
        with Chem.SDMolSupplier(sdf_file_path) as suppl:
            processed_mols = 0
            for mol in suppl:
                if mol is not None:
                    try:
                        qed_scores.append(QED.qed(mol))
                        processed_mols += 1
                    except Exception as e:
                        print(f"  - Error calculating QED for a molecule in {os.path.basename(sdf_file_path)}: {e}")
        print(f" Processed {processed_mols} molecules from {os.path.basename(sdf_file_path)}")
    except Exception as e:
        print(f" Critical error reading SDF file {sdf_file_path}: {e}")
    
    return qed_scores

def process_all_datasets(datasets_dict: dict, base_dir: str) -> pd.DataFrame:
    """
    Process all datasets to calculate QED scores and compile them into a DataFrame.
    
    Args:
        datasets_dict (dict): The dictionary defining datasets and their relative SDF file paths.
        base_dir (str): The absolute path to the directory where the script is located.

    Returns:
        A pandas DataFrame ready for plotting.
    """
    plot_data = []
    print(" Starting QED Data Processing...")
    print("="*50)

    for dataset_name, file_list in datasets_dict.items():
        print(f"\n Processing group: {dataset_name}")
        group_qed_scores = []
        for relative_path in file_list:
            # Construct the absolute path to ensure the file is found
            absolute_path = os.path.join(base_dir, relative_path)
            group_qed_scores.extend(calculate_qed_for_sdf(absolute_path))

        if not group_qed_scores:
            print(f"  ->  No molecules processed for this group.")
            continue
        
        # Deconstruct the dataset name to get metadata
        parts = dataset_name.split('_')
        pocket = parts[0].upper()
        method = 'PRISM-QED' if 'PRISM' in dataset_name else 'DiffSBDD'
        nodes = f"{parts[-2]}_{parts[-1]}" # Robustly get the last two parts

        # Add each score to our master list for the DataFrame
        for qed_score in group_qed_scores:
            plot_data.append({
                'Pocket': pocket,
                'Method': method,
                'Nodes': nodes,
                'QED': qed_score
            })
    
    print("\n" + "="*50)
    print(" QED Data Processing Complete!")
    return pd.DataFrame(plot_data)

# --- Define Datasets and Paths ---

# --- FIX for NOTEBOOKS ---
# Use os.getcwd() to get the current working directory, which is standard for notebooks.
# This assumes your notebook file is in the root directory alongside the data folders.
SCRIPT_DIRECTORY = os.getcwd()

datasets = datasets

# --- Main Execution ---

# 1. Process all SDFs and create the DataFrame
qed_dataframe = process_all_datasets(datasets, SCRIPT_DIRECTORY)

if qed_dataframe.empty:
    print("\n No data was loaded. Cannot generate plots. Please check file paths and SDF contents.")
else:
    print(f"\n DataFrame created with {len(qed_dataframe)} data points. Ready for plotting!")
    print(qed_dataframe.head())

    # --- Part 2: Violin Plot Generation ---
    print("\n Generating Violin Plots...")

    unique_pockets = sorted(qed_dataframe['Pocket'].unique())
    method_palette = {"PRISM-QED": "skyblue", "DiffSBDD": "lightgreen"}

    for pocket in unique_pockets:
        pocket_data = qed_dataframe[qed_dataframe['Pocket'] == pocket]
        node_configs = sorted(pocket_data['Nodes'].unique())
        
        fig, axes = plt.subplots(
            nrows=1, 
            ncols=len(node_configs), 
            figsize=(6 * len(node_configs), 7), 
            squeeze=False
        )
        
        fig.suptitle(f'QED Score Distribution for Pocket: {pocket}', fontsize=16, fontweight='bold')
        
        for i, nodes in enumerate(node_configs):
            ax = axes[0, i]
            plot_subset = pocket_data[pocket_data['Nodes'] == nodes]
            
            sns.violinplot(data=plot_subset, x='Method', y='QED', palette=method_palette, ax=ax)
            
            heavy_atoms = nodes.split('_')[0]
            ax.set_title(f'{heavy_atoms} Heavy Atoms', fontsize=14)
            ax.set_xlabel('Generation Method', fontsize=12)
            ax.set_ylabel('QED Score', fontsize=12)
            ax.set_ylim(0, 1)
            ax.grid(axis='y', linestyle='--', alpha=0.7)

        plt.tight_layout(rect=[0, 0, 1, 0.95])
        output_filename = f'QED_violin_plot_{pocket}.png'
        plt.savefig(output_filename, dpi=300)
        print(f"  ->  Saved plot: {output_filename}")
        plt.show()

    print("\n All plots generated successfully!")

    # --- Part 3: Summary Statistics Table Generation ---
    print("\n Generating Summary Statistics Tables...")

    # --- Table for 20 Heavy Atoms ---
    print("\n" + "="*55)
    print("        Summary Statistics for 20 Heavy Atoms")
    print("="*55)
    
    # Filter for 20 nodes
    data_20_nodes = qed_dataframe[qed_dataframe['Nodes'] == '20_nodes']
    
    if not data_20_nodes.empty:
        # Group by Pocket and Method, then calculate aggregate stats and round them
        summary_20 = data_20_nodes.groupby(['Pocket', 'Method'])['QED'].agg(['mean', 'median', 'max', 'std']).round(3)
        print(summary_20)
    else:
        print("No data available for 20 heavy atoms.")

    # --- Table for 30 Heavy Atoms ---
    print("\n" + "="*55)
    print("        Summary Statistics for 30 Heavy Atoms")
    print("="*55)
    
    # Filter for 30 nodes
    data_30_nodes = qed_dataframe[qed_dataframe['Nodes'] == '30_nodes']

    if not data_30_nodes.empty:
        # Group by Pocket and Method, then calculate aggregate stats and round them
        summary_30 = data_30_nodes.groupby(['Pocket', 'Method'])['QED'].agg(['mean', 'median', 'max', 'std']).round(3)
        print(summary_30)
    else:
        print("No data available for 30 heavy atoms.")

```

```python
import os
from rdkit import Chem
from rdkit.Chem import Draw, QED, AllChem
from IPython.display import display

def get_top_qed_molecules(sdf_file_paths: list, top_n: int = 10):
    """
    Reads molecules from SDF files, calculates their QED, and returns the top N molecules.
    """
    molecules_with_qed = []
    print(f"\nProcessing files: {', '.join([os.path.basename(p) for p in sdf_file_paths])}")
    
    for sdf_path in sdf_file_paths:
        if not os.path.exists(sdf_path):
            print(f" Warning: File not found, skipping: {sdf_path}")
            continue
        
        try:
            with Chem.SDMolSupplier(sdf_path) as suppl:
                for mol in suppl:
                    if mol is not None:
                        try:
                            qed_score = QED.qed(mol)
                            # Store the molecule and its QED score
                            molecules_with_qed.append((mol, qed_score))
                        except Exception:
                            # Skip molecules where QED calculation fails
                            continue
        except Exception as e:
            print(f" Error reading file {sdf_path}: {e}")

    # Sort the list by QED score in descending order
    molecules_with_qed.sort(key=lambda x: x[1], reverse=True)
    
    # Return the top N molecules and their scores
    return molecules_with_qed[:top_n]

# --- Define Paths ---
# This assumes your notebook is in the project's root directory
BASE_DIRECTORY = os.getcwd()

# --- NOTE: Ensure the 'datasets' dictionary is defined in a previous cell ---
# Example definition:
# datasets = {
#     '6cm4_PRISM_QED_20_nodes': ['qed_model_e72_results/drd2_test_QED_e72_model_6cm4_20_1.sdf', ...],
#     ...
# }


# --- Define Reference Molecule ---
reference_smiles = "CC1=C(C(=O)N2CCCCC2=N1)CCN3CCC(CC3)C4=NOC5=C4C=CC(=C5)F"
ref_mol = Chem.MolFromSmiles(reference_smiles)
ref_legend = "Reference: Invalid SMILES"
if ref_mol:
    AllChem.Compute2DCoords(ref_mol)
    ref_qed = QED.qed(ref_mol)
    ref_legend = f"Reference QED: {ref_qed:.3f}"

# --- Create directory for output images ---
output_dir = os.path.join(BASE_DIRECTORY, "top_qed_molecule_images")
os.makedirs(output_dir, exist_ok=True)


# --- Main Execution ---
# We will iterate through the specific datasets we want to visualize
datasets_to_display = [
    '6cm4_PRISM_QED_20_nodes',
    '6cm4_DiffSBDD_20_nodes',
    '6cm4_PRISM_QED_30_nodes',
    '6cm4_DiffSBDD_30_nodes'
]

for dataset_name in datasets_to_display:
    # Get the list of relative file paths
    relative_paths = datasets[dataset_name]
    # Create absolute paths
    absolute_paths = [os.path.join(BASE_DIRECTORY, p) for p in relative_paths]
    
    # Get the top 9 molecules to make space for the reference molecule
    top_molecules = get_top_qed_molecules(absolute_paths, top_n=9)
    
    if top_molecules:
        # Prepare for drawing
        mols_to_draw = [mol for mol, qed in top_molecules]
        legends = [f"QED: {qed:.3f}" for mol, qed in top_molecules]
        
        # --- Convert 3D molecules to 2D for display ---
        for mol in mols_to_draw:
            AllChem.Compute2DCoords(mol)
        
        # --- Add reference molecule to the list for comparison ---
        if ref_mol:
            mols_to_draw.insert(0, ref_mol)
            legends.insert(0, ref_legend)
        
        # Create a title for the plot
        title = f"Top High-QED Molecules for {dataset_name.replace('_', ' ').title()}"
        print(f"\n{title}")
        
        # Generate the grid image with higher quality
        img = Draw.MolsToGridImage(
            mols_to_draw, 
            molsPerRow=5, 
            subImgSize=(400, 400), # Increased size for higher quality
            legends=legends
        )
        
        # --- Save the image to a file ---
        # In a notebook, MolsToGridImage may return an IPython.display.Image object
        # which does not have a .save() method. We write its raw data instead.
        filename = f"{dataset_name}_top_qed.png"
        filepath = os.path.join(output_dir, filename)
        with open(filepath, 'wb') as f:
            f.write(img.data)
        print(f" Image saved to: {filepath}")
        
        # Display the image in the notebook
        display(img)
    else:
        print(f"\nNo molecules found for {dataset_name}.")

```

```python
# -*- coding: utf-8 -*-
import os, re, glob, gzip, random
from pathlib import Path
import pandas as pd
from rdkit import Chem, RDLogger
from rdkit.Chem import AllChem
from rdkit.Chem.SaltRemover import SaltRemover
from posebusters import PoseBusters

# ---------------- your datasets ----------------
datasets = {
    # 7E2Z - PRISM-QED
    '7e2z_PRISM_QED_30_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_maximum_sdfs_and_OG/QED_trained_e72_cont_50_1.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_maximum_sdfs_and_OG/QED_trained_e72_cont_50_2.sdf'
    ],
    '7e2z_PRISM_QED_20_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/QED_e72_7e2z_20_nodes.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/QED_e72_7e2z_20_nodes_2.sdf'
    ],

    # 7E2Z - DiffSBDD
    '7e2z_DiffSBDD_30_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/original_diffsbdd/7e2z_test_30_nodes_100.sdf'
    ],
    '7e2z_DiffSBDD_20_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/original_diffsbdd/7e2z_test_20_nodes_100.sdf'
    ],

    # 6CM4 - PRISM-QED
    '6cm4_PRISM_QED_20_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/drd2_test_QED_e72_model_6cm4_20_1.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/drd2_test_QED_e72_model_6cm4_20_2.sdf'
    ],
    '6cm4_PRISM_QED_30_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/drd2_test_QED_e72_model_6cm4_30_nodes.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/drd2_test_QED_e72_model_6cm4_30_nodes_2.sdf'
    ],

    # 6CM4 - DiffSBDD
    '6cm4_DiffSBDD_20_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6cm4_20_nodes_1.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6cm4_20_nodes_2.sdf'
    ],
    '6cm4_DiffSBDD_30_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6cm4_30_nodes.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6cm4_30_nodes_2.sdf'
    ],

    # 6V0U - PRISM-QED, DiffSBDD (only 20 nodes, excluded below)
    '6v0u_PRISM_QED_20_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/QED_e72_drd2_trained_6v0u_BRDtest_20_nodes.sdf'
    ],
    '6v0u_DiffSBDD_20_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6v0u_BRDtest_20_nodes.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6v0u_BRDtest_20_nodes_2.sdf'
    ],

    # 6LUQ - PRISM-QED
    '6luq_PRISM_QED_20_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/QED_e72_drd2_trained_6luq_test_20_nodes.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/QED_e72_drd2_trained_6luq_test_20_nodes_2.sdf'
    ],
    '6luq_PRISM_QED_30_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/QED_e72_drd2_trained_6luq_test_30_nodes_1.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/QED_e72_drd2_trained_6luq_test_30_nodes_2.sdf'
    ],

    # 6LUQ - DiffSBDD
    '6luq_DiffSBDD_20_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6luq_test_20_nodes_1.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6luq_test_20_nodes_2.sdf'
    ],
    '6luq_DiffSBDD_30_nodes': [
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6luq_test_30_nodes_1.sdf',
        '/Users/sanazkazeminia/Documents/mol_test_suite/data/qed_model_drd2_case_e72_results_comparisons/DiffSBDD_6luq_test_30_nodes_2.sdf'
    ]
}

# optional SuCOS 30-node file for 7E2Z only
PRISM_SUCOS_7E2Z_30 = "data/sucos_aripip_lr1e-5_ema_mean/test_time_last_ckpt_gen_100_30nodes/SuCOS_7e2z_centered_lr_1e-5_clip_0.1_30_nodes.sdf"

# ---------------- config ----------------
POCKETS_ALL = ["7E2Z", "6CM4", "6LUQ"]  # exclude 6V0U entirely, only 20 nodes exist
METHODS_MAIN = ["DiffSBDD", "PRISM_QED"]  # main cross-pocket comparison
METHODS_WITH_SUCOS = ["DiffSBDD", "PRISM_QED", "PRISM_SuCOS"]  # 7E2Z view
CAP_PER_POCKET_METHOD = 100
OUTPUT_DIR = "posebusters_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)
SEED = 42

# ---------------- RDKit salvage utilities ----------------
saltremover = SaltRemover()
RDLogger.DisableLog('rdApp.*')

def add_nitrogen_charges(mol: Chem.Mol) -> Chem.Mol:
    """Fix common tertiary amine cases where N has valence 4 and missing +1 charge."""
    mol.UpdatePropertyCache(strict=False)
    probs = Chem.DetectChemistryProblems(mol)
    if not probs:
        Chem.SanitizeMol(mol)
        return mol
    for p in probs:
        if p.GetType() == 'AtomValenceException':
            at = mol.GetAtomWithIdx(p.GetAtomIdx())
            if at.GetAtomicNum() == 7 and at.GetFormalCharge() == 0 and at.GetExplicitValence() == 4:
                at.SetFormalCharge(1)
    Chem.SanitizeMol(mol)
    return mol

# optional OpenBabel pH correction, used as a last resort
try:
    from openbabel import openbabel as ob
    def correct_for_pH(mol: Chem.Mol, pH: float = 7.4) -> Chem.Mol | None:
        try:
            obmol = ob.OBMol()
            conv = ob.OBConversion()
            conv.SetInFormat("pdb")
            if not mol.HasProp('_Name'):
                mol.SetProp('_Name', 'molecule')
            conv.ReadString(obmol, Chem.MolToPDBBlock(mol))
            obmol.CorrectForPH(pH)
            conv.SetOutFormat("mol")
            neo = Chem.MolFromMolBlock(conv.WriteString(obmol).strip(), sanitize=True)
            return neo
        except Exception:
            return None
    HAVE_OBABEL = True
except Exception:
    HAVE_OBABEL = False

def salvage_mol(raw: Chem.Mol) -> Chem.Mol | None:
    """Strip salts, repair missing nitrogen charges, optional pH correction if needed."""
    if raw is None:
        return None
    try:
        m = saltremover.StripMol(raw, dontRemoveEverything=True)
        m = add_nitrogen_charges(m)
        return m
    except Exception:
        if HAVE_OBABEL:
            try:
                neo = correct_for_pH(raw)
                if neo is not None:
                    return add_nitrogen_charges(neo)
            except Exception:
                pass
        return None

# ---------------- helpers ----------------
def parse_key(k: str):
    # keys like '7e2z_PRISM_QED_30_nodes'
    m = re.match(r'(?i)^(7e2z|6cm4|6luq|6v0u)_(DiffSBDD|PRISM_QED|PRISM_SuCOS)_(\d+)_nodes$', k)
    if not m:
        return None
    pocket, method, nodes = m.group(1).upper(), m.group(2), int(m.group(3))
    return pocket, method, nodes

def build_entries_from_datasets(ds: dict, want_nodes=30, exclude_pockets=("6V0U",)):
    """Return dict[method] = list of {'path', 'pocket'} using dataset keys to decide node count."""
    out = {"DiffSBDD": [], "PRISM_QED": [], "PRISM_SuCOS": []}
    for k, paths in ds.items():
        parsed = parse_key(k)
        if not parsed:
            continue
        pocket, method, nodes = parsed
        if pocket in exclude_pockets:
            continue
        if nodes != want_nodes:
            continue
        for p in paths:
            out[method].append({"path": p, "pocket": pocket})
    # add SuCOS file if it exists, 7E2Z only
    if os.path.isfile(PRISM_SUCOS_7E2Z_30):
        out["PRISM_SuCOS"].append({"path": PRISM_SUCOS_7E2Z_30, "pocket": "7E2Z"})
    return out

def iter_sdf_mols(path: str):
    """Yield RDKit mols from .sdf or .sdf.gz without initial sanitization."""
    p = Path(path)
    try:
        if p.suffix.lower() == ".gz":
            with gzip.open(path) as gzh:
                supplier = Chem.ForwardSDMolSupplier(gzh, sanitize=False, removeHs=False)
                for m in supplier:
                    yield m
        else:
            supplier = Chem.ForwardSDMolSupplier(str(p), sanitize=False, removeHs=False)
            for m in supplier:
                yield m
    except Exception:
        return

def load_posebusters_rows(entries_for_method, method_name, cap=CAP_PER_POCKET_METHOD, seed=SEED):
    """
    entries_for_method: list of {'path', 'pocket'}
    returns rows and QC stats
    """
    rng = random.Random(seed)
    by_pocket = {}
    for e in entries_for_method:
        by_pocket.setdefault(e["pocket"], []).append(e["path"])

    rows = []
    qc_stats = {}  # (pocket, method)  dict of counts

    for pocket, file_list in by_pocket.items():
        raw_read = 0
        salvaged = 0
        dropped = 0
        seen_smiles = set()
        pool = []

        for fp in file_list:
            if not os.path.isfile(fp):
                continue
            for raw in iter_sdf_mols(fp):
                raw_read += 1
                m = salvage_mol(raw)
                if m is None:
                    dropped += 1
                    continue
                smi = Chem.MolToSmiles(m, isomericSmiles=True)
                if smi in seen_smiles:
                    continue
                seen_smiles.add(smi)
                pool.append(m)
                salvaged += 1

        rng.shuffle(pool)
        pool = pool[:cap]
        for m in pool:
            rows.append({"mol": m, "Pocket": pocket, "Method": method_name})

        qc_stats[(pocket, method_name)] = {
            "RawRead": raw_read,
            "SalvagedKept": len(pool),
            "DroppedBeforeCap": max(0, raw_read - salvaged),
            "DedupedDropped": max(0, salvaged - len(pool)),
        }

    return rows, qc_stats

def run_posebusters_for_method(method_name, entries_for_method, config="mol_fast"):
    if not entries_for_method:
        print(f"[{method_name}] No SDF files at desired node count.")
        return pd.DataFrame(), {}

    rows, qc = load_posebusters_rows(entries_for_method, method_name)

    if not rows:
        print(f"[{method_name}] No molecules loaded after salvage and capping.")
        return pd.DataFrame(), qc

    buster = PoseBusters(config=config)
    out_frames = []
    for pocket in sorted({r["Pocket"] for r in rows}):
        mols = [r["mol"] for r in rows if r["Pocket"] == pocket]
        if not mols:
            continue
        df = buster.bust(mols)
        df["Pocket"] = pocket
        df["Method"] = method_name
        out_frames.append(df)
    return (pd.concat(out_frames, ignore_index=True) if out_frames else pd.DataFrame()), qc

def pretty_table(df, pockets, methods):
    df = df.copy()
    df = df.reindex(index=pockets, columns=methods)
    # avoid FutureWarning, use Series.map on each column
    return df.apply(lambda col: col.map(lambda x: "" if pd.isna(x) else f"{x:.1f}"))

# ---------------- run ----------------
METHOD_ENTRIES = build_entries_from_datasets(datasets, want_nodes=30, exclude_pockets=("6V0U",))

all_frames = []
all_qc = {}

# run DiffSBDD, PRISM_QED, and optional PRISM_SuCOS
for method in ["DiffSBDD", "PRISM_QED", "PRISM_SuCOS"]:
    print(f" {method}")
    df, qc = run_posebusters_for_method(method, METHOD_ENTRIES.get(method, []), config="mol_fast")
    if not df.empty:
        all_frames.append(df)
    all_qc.update(qc)

if not all_frames:
    raise SystemExit("No PoseBusters results produced after salvage.")

raw = pd.concat(all_frames, ignore_index=True)

# determine pass/fail
bool_cols = [c for c in raw.columns if raw[c].dtype == bool]
if not bool_cols:
    raise SystemExit("No boolean PoseBusters checks found, cannot compute pass rates.")
raw["passed_all_checks"] = raw[bool_cols].all(axis=1)

# aggregate
agg = (raw.groupby(["Pocket", "Method"])
          .agg(Total=("passed_all_checks","size"),
               Passed=("passed_all_checks","sum"))
          .reset_index())
agg["Pass Rate (%)"] = (agg["Passed"] / agg["Total"] * 100)

# ----- Table A, cross-pocket, 30-node, DiffSBDD vs PRISM_QED -----
A = (agg[agg["Method"].isin(METHODS_MAIN)]
        .set_index(["Pocket","Method"])["Pass Rate (%)"]
        .unstack("Method"))
A = A.reindex(index=POCKETS_ALL, columns=METHODS_MAIN)

A_fmt = pretty_table(A, POCKETS_ALL, METHODS_MAIN)

print("\n=== PoseBusters pass rate (%)  30-node, capped at 100 per pocketmethod ===")
print(A_fmt.to_string())

# counts actually used, Table A methods only
counts_A = (agg[agg["Method"].isin(METHODS_MAIN)]
            .set_index(["Pocket","Method"])["Total"].unstack("Method"))
counts_A = counts_A.reindex(index=POCKETS_ALL, columns=METHODS_MAIN)
print("\n(molecule counts actually used per pocketmethod, Table A)")
print(counts_A.fillna(0).astype(int).to_string())

# ----- Table B, 7E2Z only, add PRISM_SuCOS -----
B = (agg[(agg["Pocket"]=="7E2Z") & (agg["Method"].isin(METHODS_WITH_SUCOS))]
        .set_index(["Pocket","Method"])["Pass Rate (%)"]
        .unstack("Method"))
B = B.reindex(index=["7E2Z"], columns=METHODS_WITH_SUCOS)
B_fmt = pretty_table(B, ["7E2Z"], METHODS_WITH_SUCOS)

print("\n=== 7E2Z only, adding PRISM_SuCOS (similarity-conditioned generation) ===")
print(B_fmt.to_string())

counts_B = (agg[(agg["Pocket"]=="7E2Z") & (agg["Method"].isin(METHODS_WITH_SUCOS))]
            .set_index(["Pocket","Method"])["Total"].unstack("Method"))
counts_B = counts_B.reindex(index=["7E2Z"], columns=METHODS_WITH_SUCOS)
print("\n(molecule counts for 7E2Z view)")
print(counts_B.fillna(0).astype(int).to_string())

# ----- QC table with salvage diagnostics -----
qc_records = []
for (pocket, method), stats in all_qc.items():
    if pocket not in POCKETS_ALL and not (pocket == "7E2Z" and method == "PRISM_SuCOS"):
        continue
    row = {"Pocket": pocket, "Method": method}
    row.update(stats)
    qc_records.append(row)

qc_df = pd.DataFrame(qc_records)
qc_df = qc_df.sort_values(["Pocket","Method"])
print("\n[QC] Raw read, salvaged and deduped counts per pocketmethod")
print(qc_df.to_string(index=False))

# # ----- save outputs -----
# os.makedirs(OUTPUT_DIR, exist_ok=True)
# raw.to_csv(os.path.join(OUTPUT_DIR, "posebusters_raw_with_checks.csv"), index=False)
# A.to_csv(os.path.join(OUTPUT_DIR, "posebusters_passrate_tableA_numeric.csv"))
# A_fmt.to_csv(os.path.join(OUTPUT_DIR, "posebusters_passrate_tableA_pretty.csv"))
# counts_A.to_csv(os.path.join(OUTPUT_DIR, "posebusters_counts_tableA.csv"))
# B.to_csv(os.path.join(OUTPUT_DIR, "posebusters_passrate_tableB_numeric.csv"))
# B_fmt.to_csv(os.path.join(OUTPUT_DIR, "posebusters_passrate_tableB_pretty.csv"))
# counts_B.to_csv(os.path.join(OUTPUT_DIR, "posebusters_counts_tableB.csv"))
# qc_df.to_csv(os.path.join(OUTPUT_DIR, "posebusters_qc_salvage_counts.csv"), index=False)

print("\nDone, outputs written to:", OUTPUT_DIR)
```
